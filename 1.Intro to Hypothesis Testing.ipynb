## Introduction to Hypothesis Testing

### What is hypothesis Testing?

Hypothesis testing is a statistical method, that can be used in ML to make inferences about population based on sample of data.
In Hypothesis testing, a hypothesis is formulated about the population, and the goal is to determine whether the available evidence from the sample supports or contradicts the sample.

### What is Null Hypothesis (Ho)?

In statistics, the null hypothesis is a statement that there is no significant difference between two groups or no significant relationship between two variables. The null hypothesis assumes that any observed difference or relationship is due to chance or random variation in the data.
It provides a baseline against which to compare the observed data. If the observed data significantly deviates from the null hypothesis, we can reject the null hypothesis and conclude that there is a statistically significant difference or relationship between the variables being tested.
The null hypothesis is the statement that we assume to be true initially. It is usually the opposite of what we want to prove, and it is denoted by H0.

### What is alternative hypothesis (Ha)?

In statistics, the alternative hypothesis is a statement that contradicts the null hypothesis and asserts that there is a significant difference or relationship between two groups or variables. It represents the possibility that any observed difference or relationship is not due to chance but is instead caused by some real effect.
The alternative hypothesis is the statement we want to prove. It is denoted by Ha and is usually the opposite of the null hypothesis.

### What is Hypothesized Value?

A hypothesized value, also known as a null hypothesis value, is a value or condition that is assumed to be true for the purpose of testing a statistical hypothesis. 

The hypothesized value serves as a reference point for evaluating the statistical significance of the observed data.
The hypothesized value is used to determine the likelihood of the observed data being due to random chance or to some real effect.

### What is Actual value?

Is the value that is observed or measured in the sample data. This value is the result of collecting data from a sample and is used to determine whether the hypothesized value is likely to be true or false.

### Difference between the hypothesized value and actual value?

The difference between the hypothesized value and actual value is an important factor in hypothesis testing, as it influences the test statistic and the outcome of the test.

The sample size, the level of significance, and the variability of the data also play important roles in determining the results of the test.

### What are the Type-I and Type-II Errors? When they will Occur in Hypothesis Testing?

Type 1 and Type 2 errors occur in hypothesis testing when we make a decision about a population parameter based on a sample of data.

Type 1 error occurs (false positive) This occurs when the null hypothesis is rejected even though it is true. In other words, we conclude that there is a significant difference or effect in the population, when in fact there is no such difference or effect. 

The probability of making a Type 1 error is denoted by the symbol alpha (α), and it is typically set to 0.05 or 0.01. If the alpha level is set at 0.05, then there is a 5% chance of committing a Type 1 error.

Type 2 error occurs (false negative) when we fail to reject a null hypothesis that is actually false. In other words, we conclude that there is no significant difference or effect in the population, when in fact there is such a difference or effect. 

The probability of making a Type 2 error is denoted by the symbol beta (β). The power of a test is equal to 1-β, which represents the probability of correctly rejecting a false null hypothesis. The power of a test is affected by factors such as sample size, the effect size, and the level of significance.

In hypothesis testing, we aim to control both Type 1 and Type 2 errors by choosing an appropriate level of significance (alpha), calculating the power of the test, and interpreting the results in the context of the problem at hand. A well-designed hypothesis test balances the risk of making a Type 1 error (rejecting a null hypothesis that is true) with the risk of making a Type 2 error (failing to reject a null hypothesis that is false).

Both Type 1 and Type 2 errors can be caused by various factors such as inadequate sample size, flawed study design, biased data collection or analysis, and random chance.

Type 1 errors can also occur when the significance level (alpha) is set too high, which increases the likelihood of rejecting the null hypothesis even when it is true. On the other hand, Type 2 errors can occur when the statistical power of the test is too low, which increases the likelihood of failing to reject the null hypothesis even when it is false.
